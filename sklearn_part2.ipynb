```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
carsales=pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/car-sales-extended-missing-data.csv")
carsales.head(10)
```

```python
carsales.isna().sum()
```

```python
x=carsales.drop("Price",axis=1)
y=carsales["Price"]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
categorial_feature=["Make","Colour","Doors"]
one_hot=OneHotEncoder()
transformer=ColumnTransformer([("one_hot",one_hot,categorial_feature)],remainder="passthrough")
transformed_x=transformer.fit_transform(x)
transformed_x
```

```python
carsales["Make"].fillna("missing",inplace=True)
carsales["Colour"].fillna("missing",inplace=True)
carsales["Odometer (KM)"].fillna(carsales["Odometer (KM)"].mean(),inplace=True)
carsales["Doors"].fillna(4,inplace=True)
```

```python
carsales.isna().sum()
```

```python
carsales.dropna(inplace=True)
```

```python
carsales.isna().sum()
```

```python
x=carsales.drop("Price",axis=1)
y=carsales["Price"]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
categorial_feature=["Make","Colour","Doors"]
one_hot=OneHotEncoder()
transformer=ColumnTransformer([("one_hot",one_hot,categorial_feature)],remainder="passthrough")
transformed_x=transformer.fit_transform(carsales)
transformed_x
```

```python
carsalesmissing=pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/car-sales-extended-missing-data.csv")
```

```python
carsalesmissing
```

```python
carsalesmissing.isna().sum()
```

```python
carsalesmissing.dropna(subset=["Price"],inplace=True)
carsalesmissing.isna().sum()
```

```python
x=carsalesmissing.drop("Price",axis=1)
y=carsalesmissing["Price"]
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
cat_imputer=SimpleImputer(strategy="constant",fill_value="missing")
door_imputer=SimpleImputer(strategy="constant", fill_value=4)
num_imputer=SimpleImputer(strategy="mean")

cat_feature=["Make","Colour"]
door_feature=["Doors"]
num_feature=["Odometer (KM)"]

imputer=ColumnTransformer([
                           ("cat_imputer",cat_imputer,cat_feature),
                           ("door_imputer",door_imputer,door_feature),
                           ("num_inputer",num_imputer,num_feature)
])
filled_x=imputer.fit_transform(x)
filled_x
```

```python
carsalesfilled=pd.DataFrame(filled_x,columns=["Make","Colours","Doors","Odometer (km)"])
carsalesfilled
```

```python
carsalesfilled.isna().sum()
```

```python
x2=carsalesfilled
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
categorial_feature=["Make","Colours","Doors"]
one_hot=OneHotEncoder()
transformer=ColumnTransformer([("one_hot",one_hot,categorial_feature)],remainder="passthrough")
transformed_x=transformer.fit_transform(x2)
transformed_x
```

```python
np.random.seed(42)
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(transformed_x,y,test_size=0.2)
model=RandomForestRegressor()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```
